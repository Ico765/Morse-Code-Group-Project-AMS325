```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("touch_duration_report.csv")

print(data.head())
```

```python
plt.figure(figsize=(8, 6))
plt.hist(data["Touch Duration"], bins=20, color='skyblue', edgecolor='black')
plt.title("Touch Duration Distribution")
plt.xlabel("Touch Duration (seconds)")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
input_counts = data["Input Type"].value_counts()

plt.figure(figsize=(6, 6))
input_counts.plot(kind='bar', color=['lightgreen', 'orange'])
plt.title("Short vs Long Touch Count")
plt.xlabel("Input Type")
plt.ylabel("Count")
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()
```

```python
success_data = data[data["Input Type"] == "check_password"]["Success/Fail"].value_counts()

plt.figure(figsize=(6, 6))
success_data.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['lightblue', 'salmon'])
plt.title("Password Match Success Rate")
plt.ylabel("")  
plt.show()
```

```python
import pandas as pd

valid_touch_data = data[data["Touch Duration"] > 0]

user_avg_durations = valid_touch_data.groupby("User ID")["Touch Duration"].mean()

plt.figure(figsize=(10, 6))
user_avg_durations.plot(kind='bar', color='purple')
plt.title("Average Touch Duration by User")
plt.xlabel("User ID")
plt.ylabel("Average Touch Duration (seconds)")
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```
